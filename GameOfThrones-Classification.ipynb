```python
import pandas as pd
df=pd.read_csv("battles.csv")
df
```

```python
from sklearn.preprocessing import LabelEncoder #classifying the data 
enc=LabelEncoder() 
enc.fit(df.attacker_1) #fitting the color attribute i.e mapping of i/p & o/p variables
df.attacker_1=enc.transform(df.attacker_1) #conversion from string to numeric
enc.fit(df.defender_1) #fitting the color attribute i.e mapping of i/p & o/p variables
df.defender_1=enc.transform(df.defender_1) #conversion from string to numeric
enc.fit(df.location) #fitting the color attribute i.e mapping of i/p & o/p variables
df.location=enc.transform(df.location) #conversion from string to numeric
df
```
